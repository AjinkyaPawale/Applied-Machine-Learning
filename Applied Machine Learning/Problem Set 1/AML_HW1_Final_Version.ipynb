### Function for calculating median

```python
def median_price(price):
    price_nonzero = []
    median=0
    for i in range(0,len(price)):
        if price[i] != 0:
            price_nonzero.append(price[i])
    price_nonzero.sort()    
    n=len(price_nonzero)
    if n % 2 == 0: 
        median1 = price_nonzero[n//2] 
        median2 = price_nonzero[n//2 - 1] 
        median = (median1 + median2)/2
    else: 
        median = price_nonzero[n//2] 
    return median
```

### Function for correlation coefficient

```python
# Ref: https://www.questionpro.com/blog/pearson-correlation-coefficient/
def pearson_corr(x, y):
    n=len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x_sq = sum(xj*xj for xj in x)
    sum_y_sq = sum(yj*yj for yj in y)
    sum_xy = sum(xj*yj for xj,yj in zip(x,y))
    num = sum_xy - (sum_x*sum_y/n)
    den = pow((sum_x_sq-pow(sum_x,2)/n)*(sum_y_sq-pow(sum_y,2)/n),0.5)
    if den == 0: return 0
    return num/den
```

### Correlation coefficients for original data set

### One hot encoding can be applied to the categorical values which are nominal i.e. in this case city and statezip. For categorical values like bedroom, bathrooms, floors, view, waterfront and condition, label encoding works because they are ordinal. 

### The below example one hot encodes city and statezip with the condition that the frequencies greater than 100 will form one hot vectors and the rest will fall under "others" category. This is to ensure that we don't have many one hot vectors of city and statezip unnecessarily.

### The below example consists of a threshold, which is applied to the frequencies of statezip and city, so that there are a limited number of one hot encoded vectors to deal with.

### For Data Original

```python
#ref:https://stackoverflow.com/questions/44527731/one-hot-encoding-for-frequent-values-only
#ref:https://stackoverflow.com/questions/37265312/how-to-create-dummies-for-certain-columns-by-pandas-get-dummies-method
#ref:https://queirozf.com/entries/one-hot-encoding-a-feature-on-a-pandas-dataframe-an-example

threshold = 100
counts_city = data.city.value_counts()
counts_statezip = data.statezip.value_counts()
repl_city = counts_city[counts_city <= threshold].index
repl_statezip = counts_statezip[counts_statezip <= threshold].index
one_hot_data = pd.concat([data,pd.get_dummies(data.city.replace(repl_city, 'others'),prefix='city')
                          ,pd.get_dummies(data.statezip.replace(repl_statezip, 'others'),prefix='statezip')],axis=1).drop(['city',"statezip"],axis=1)
one_hot_data.head()
```

### Functions to calculate mean absolute error and root mean square error

```python
import math 
def MAE(y_pred,y_actual):
    return (sum(abs(y_pred-y_actual)) / len(y_pred))

def RMSE(y_pred,y_actual):
    return (math.sqrt(sum(np.square(y_pred-y_actual)) / len(y_pred)))
```

### Regardless of normalizing or standardizing, the replacement of zeros with median or mean improved the error in all cases (MAE and RMSE). 
### For elimiating instances with zeros, when measuring MAE, the result improved. This is because when calculating the MAE in the original dataset, instances with zeros drastically increase the MAE (values with zero have an MAE of the mean). When measuring the RMSE for instances with replaced zeros, the error increased. Unlike the MAE, which divides the error by N (number of instances), RMSE divides the results by the square root of N, which emphasizes the effect of errors. Since when removing instances with zeros increases the error, it is likely that the instances that had zero in price actually were really good examples of the dataset except for the price attribute.
